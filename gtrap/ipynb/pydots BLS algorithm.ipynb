```python
import pydotplus as pydot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
#graph = pydot.Dot(graph_type='graph') #w/o arrows
```

```python
dataset="Light Curve (LC)"
stack="Stack the residual in BLS spectrum (P)"
bf=[]
cr=[]
binlc=[]
blsspec="BLS spectrum"
nb=3
nbb=2
tag=["1","2","3"]
ttag=["1","2"]
for i in range(0,nb):
    for j in range(0,nbb):
        bf.append("Thread"+tag[i])
        cr.append("Residuals for P=P"+ttag[j])
        binlc.append("Binned LC for P=P"+ttag[j])

graph = pydot.Dot(graph_type='digraph',dpi=150) #w arrows


### node definition
node = pydot.Node(name=dataset,color="red")
graph.add_node(node)
node = pydot.Node(name=blsspec,color="red")
graph.add_node(node)
for i in binlc:
    node = pydot.Node(name=i,shape="box",color="blue")
    graph.add_node(node)
for i in cr:
    node = pydot.Node(name=i,shape="box",color="blue")
    graph.add_node(node)

### making edges
k=0
for i in range(0,nb):
    for j in range(0,nbb):
        if i==0:
            edge = pydot.Edge(dataset,binlc[k],label="Block"+tag[j]+": Folding \n LC with P=P"+tag[j], color="red", style="dashed")            
            graph.add_edge(edge)
            edge = pydot.Edge(cr[k],blsspec,label="Minimum", color="red", style="dashed")
            graph.add_edge(edge)
        edge = pydot.Edge(binlc[k], cr[k],label=bf[k],color="blue")
        graph.add_edge(edge)    
        #edge = pydot.Edge(cr[k],binlc[k])
        #graph.add_edge(edge)
        k=k+1
```

```python
filename="example1.png"
graph.write_png(filename)
img = mpimg.imread(filename)
fig=plt.figure(figsize=(40,40))
plt.imshow(img)
plt.show()
filename="example1.pdf"
graph.write_pdf(filename)
```

```python
mm="main memory (host)"
gm="global memory"
bl=["Block1,\n Thread1,2,3...","Block2,\n Thread1,2,3..."]
sm=["shared memory1","shared memory2"]
graph = pydot.Dot(graph_type='digraph',dpi=150) #w arrows


### node definition
node = pydot.Node(name=mm,shape="ellipse",color="black")
graph.add_node(node)
node = pydot.Node(name=gm,color="red")
graph.add_node(node)

for i in bl:
    node = pydot.Node(name=i,shape="box",color="black")
    graph.add_node(node)
for i in sm:
    node = pydot.Node(name=i,shape="box",color="blue")
    graph.add_node(node)

### making edges
edge = pydot.Edge(mm,gm)           
graph.add_edge(edge)
edge = pydot.Edge(gm,mm)           
graph.add_edge(edge)
k=0
for i in bl:
    edge = pydot.Edge(i,gm,label="slow", color="red", style="dashed")            
    graph.add_edge(edge)
    edge = pydot.Edge(gm,i, color="red", style="dashed")            
    graph.add_edge(edge)
    edge = pydot.Edge(i,sm[k],label="fast", color="blue")            
    graph.add_edge(edge)
    edge = pydot.Edge(sm[k],i, color="blue")            
    graph.add_edge(edge)   
    k=k+1
```

```python
filename="example2.png"
graph.write_png(filename)
img = mpimg.imread(filename)
fig=plt.figure(figsize=(40,40))
plt.imshow(img)
plt.show()
filename="example2.pdf"
graph.write_pdf(filename)
```
