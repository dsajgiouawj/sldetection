```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import pylab
```

```python
P=2.2047299 ## period of HAT-P-7
```

```python
file="/sharksuck/kic/data/0106/010666592/kplr010666592-2009131105131_llc.fits"
hdulist = fits.open(file)
t=hdulist['LIGHTCURVE'].data.field("TIME")
flux=hdulist['LIGHTCURVE'].data.field('PDCSAP_FLUX')
```

```python
fig = plt.figure()
ax=fig.add_subplot(111)
ax.plot(t,flux)
plt.show()
```
